```python
##
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
## defining target size of output after convolution:

def target_size(img_size, kernel_size , stride =1, padding =0):
  """ For target size of convolution with:
      w - img size
      k - kernel size
      s - stride
      p -  padding
      it is given by:
      target_size =  (w - k + 2*p/s)+1"""
  target_size  =  (img_size - kernel_size + 2*padding/stride)+1
  return target_size
```

```python
# Convolve Image
def convolve_image(image):
  k = 3 # kernal size
  s = 1 # stride
  p = 0 # padding
  w = image.shape[0]
  targetSize = int(target_size(w, k, s, p))  # convert float to integer as The np.zeros function expects integer values for its shape argument

  # kernel: laplacian kernel for edge detection
  kernel =np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])

  convolved_image = np.zeros([targetSize, targetSize])
  # convolve
  for i in range(targetSize):
    for j in range(targetSize):
      mat = image[i: i+k, j : j+k]
      convolved_image[i,j] = np.sum(np.multiply(mat, kernel))
  return convolved_image
```

```python
if __name__ =='__main__':
  img = '/content/download.jpeg'
  image = cv2.imread(img, 0)
  print("Original Image: \n")
  cv2_imshow(image)
  convolved_image = convolve_image(image)
  print("Convolved Image: \n")
  cv2_imshow(convolved_image)
```
